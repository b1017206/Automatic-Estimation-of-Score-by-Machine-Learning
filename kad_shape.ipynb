```python
import glob
import os
import shutil
import zipfile
import copy
```

```python
#課題毎に変更する
kadai='kadai73'
kad='kad73'
#ja_folderも変更
```

# jarファイルの場所確認

```python
# ファイルを取得する
path = './'+kadai+'/*/*.jar'
flist = glob.glob(path)
print(flist)
```

# kadai__配下のフォルダの学生番号を抽出

```python
a_list=[]
for file in flist:
    a=file[10:14]
    a_list.append(a)
print(a_list)
```

# jarファイルをrenameとkadai__配下に移動

```python
print('変更前')
print(flist)

# ファイル名を一括で変更する
for num,file2 in enumerate(flist):
    os.rename(file2, './'+kadai+'/'+kad+'_18%s.jar' %a_list[num])

# ファイルを取得する
path2 = './'+kadai+'/*'
flist2 = glob.glob(path2)
print('変更後')
print(flist2)
```

# kadai__配下のフォルダ削除

```python
# ファイルを取得する
path3 = './'+kadai+'/????-????????_??????'
flist3 = glob.glob(path3)
print('変更前')
print(flist3)
for folder in flist3:
    shutil.rmtree(folder)
print('変更後')
# ファイルを取得する
path4 = './'+kadai+'/*'
flist4 = glob.glob(path4)
print(flist4)
```

# jarファイルからクラス(javaファイル)を取り出し、kadai__配下に置く

```python
# ファイルを取得する
path2 = './'+kadai+'/*'
flist2 = glob.glob(path2)
print(flist2)
a_list2=[]
for file2 in flist2:
    a2=file2[18:22]
    a_list2.append(a2)
print(a_list2)
```

```python
#解凍後のjavaファイル(クラスファイル)の表示
#クラスの数やファイル名が課題内容と一致しているか確認
#違うときはflist2[_]の要素を変更する
with zipfile.ZipFile(flist2[1], 'r') as zf:
    list_java=zf.namelist()
list_java
```

```python
#コピペルナー対象ファイル（上記のlist_javaより）のlist
remove_list_java=copy.copy(list_java)
#list_javaの削除する要素を入力
del remove_list_java[0]
del remove_list_java[1]
del remove_list_java[1]
del remove_list_java[1]
del remove_list_java[1]
del remove_list_java[1]
del remove_list_java[1]
#コピペルナー対象フォルダ（上記のlist_javaより）.javaの親を指定
ja_folder='db2'
remove_list_java
```

```python
#ファイル名だけ取ってくる
list_filename=[s.lstrip(ja_folder+'/') for s in list_java]
list_filename
```

```python
#jarファイルの解凍→folder配下にjavaファイル(クラス)がある状態
for num,jarname in enumerate(flist2):
    with zipfile.ZipFile(jarname, 'r') as zf:
        zf.extractall('./'+kadai+'/'+kad+'_18%s' %a_list2[num])
print(glob.glob('./'+kadai+'/*'))
```

```python
a_list3=[]
for id_name in a_list2:
    targetDirPath = './'+kadai+'/'+kad+'_18%s/%s'%(id_name,ja_folder)
    #print(targetDirPath)
    for dirPath, dirList, fileList in os.walk(targetDirPath):
        result = list(set(list_filename) - set(fileList))
        #print(result)
        if not result:#もしもresultが空なら
            a_list3.append(id_name)
print(a_list3)
```

```python
#kadai__配下に置くようにリネーム
for num,jarname in enumerate(a_list3):
    #print(num)
    for i,file in enumerate(remove_list_java):#コピペルナー対象のファイルの分だけ回す
        file_name='./'+kadai+'/'+kad+'_18%s/%s'%(a_list3[num],file)
        file_name_new='./'+kadai+'/'+kad+'_18%s_%s.java'%(a_list3[num],i)
        #print(file_name+'  '+file_name_new)
        os.rename(file_name, file_name_new)
print(glob.glob('./'+kadai+'/*'))
```

# kadai__配下の.jarファイルとフォルダの削除
